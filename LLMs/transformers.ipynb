# SENTIMENT ANALYSIS

```python
from transformers import pipeline

# Load sentiment analysis pipeline using DistilBERT
sentiment_pipeline = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
text = ["I love Kaggle!", "This competition is very tough.", "I hate waiting for results."]
results = sentiment_pipeline(text)

for t, r in zip(text, results):
    print(f"Text: {t}\nSentiment: {r['label']}, Confidence: {r['score']:.4f}\n")
```

```python
sample=['this product is good', 'not worth the money', 'poor performance', 'it was okay, not bad', 'phenomenal performance']
res=sentiment_pipeline(sample)

for t, r in zip(sample, res):
    print(f"Text: {t}\nSentiment: {r['label']}, Confidence: {r['score']:.4f}\n")
```

```python
classifier = pipeline(task="text-classification", model="SamLowe/roberta-base-go_emotions", top_k=None)
```

# TEXT SUMMARIZATION

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
text = """Kaggle is a platform for data science and machine learning competitions. 
It provides datasets, notebooks, and a community for users to collaborate, 
learn, and solve real-world problems. Many organizations host competitions 
to crowdsource solutions, and winners often receive prizes and job opportunities."""

summary = summarizer(text, max_length=50, min_length=20, do_sample=False)

print("Summary:", summary[0]['summary_text'])
```

```python
text = """
India is a country located in South Asia, known for its rich history, diverse culture, and significant influence on global affairs. It is the world's largest democracy and the second-most populous country, with over 1.4 billion people. India is a federal parliamentary republic, with 28 states and 8 Union Territories.

Geographically, India is marked by diverse landscapes, including the Himalayan mountain range to the north, vast plains, deserts, and a long coastline along the Indian Ocean. The country has a wide range of climates, from tropical in the south to temperate in the north.

India is known for its diverse cultural heritage, with a blend of ancient traditions and modern practices. Major religions include Hinduism, Islam, Christianity, Sikhism, Buddhism, and Jainism. The country is home to various languages, with Hindi and English being widely spoken, though there are 22 officially recognized languages.
"""
summarized_text = summarizer(text, min_length=5, max_length=140)[0]['summary_text']
summarized_text
```
